```python
# Data Exploration —  SafeClick 
# Explore and understand the phishing URL dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '..')
import config

# Set plot style
plt.style.use('dark_background')
sns.set_palette("Set2")
```

```python
# Load dataset
df = pd.read_csv(config.FINAL_DATASET_CSV)
print(f"Dataset shape: {df.shape}")
print(f"\nFirst 10 rows:")
df.head(10)
```

```python
# Dataset info
print("Dataset Info:")
print(f"Total URLs: {len(df)}")
print(f"Phishing URLs: {(df['label'] == 1).sum()}")
print(f"Legitimate URLs: {(df['label'] == 0).sum()}")
print(f"\nNull values:\n{df.isnull().sum()}")
```

```python
# Distribution of labels
fig, ax = plt.subplots(figsize=(6, 4))
df['label'].value_counts().plot(kind='bar', color=['#27ae60', '#e74c3c'], ax=ax)
ax.set_xticklabels(['Legitimate (0)', 'Phishing (1)'], rotation=0)
ax.set_title('Distribution of URL Labels')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# URL length analysis
df['url_length'] = df['url'].apply(len)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df[df['label']==0]['url_length'].hist(bins=30, ax=axes[0], color='#27ae60', alpha=0.7)
axes[0].set_title('Legitimate URL Lengths')
axes[0].set_xlabel('URL Length')

df[df['label']==1]['url_length'].hist(bins=30, ax=axes[1], color='#e74c3c', alpha=0.7)
axes[1].set_title('Phishing URL Lengths')
axes[1].set_xlabel('URL Length')

plt.tight_layout()
plt.show()

print(f"Average legitimate URL length: {df[df['label']==0]['url_length'].mean():.1f}")
print(f"Average phishing URL length: {df[df['label']==1]['url_length'].mean():.1f}")
```
