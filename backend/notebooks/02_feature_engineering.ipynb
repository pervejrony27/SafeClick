Define and compute features for phishing detection.

```python
# Feature Engineering — Test feature extraction

import pandas as pd
import sys
sys.path.insert(0, '..')
from ml.feature_extractor import extract_features
import config
```

```python
# Test on a single URL
url = "http://192.168.1.1/login/bank-secure/verify.php?id=123"
features = extract_features(url)

print(f"URL: {url}\n")
print(f"Total features extracted: {len(features)}\n")
for name, value in features.items():
    print(f"  {name:30s} = {value}")
```

```python
# Compare features: Safe vs Phishing
safe_url = "https://www.google.com"
phish_url = "http://192.168.1.1/login/bank-secure/verify.php"

safe_features = extract_features(safe_url)
phish_features = extract_features(phish_url)

comparison = pd.DataFrame({
    'Feature': list(safe_features.keys()),
    'Safe URL': list(safe_features.values()),
    'Phishing URL': list(phish_features.values()),
})
comparison
```

```python
# Extract features from entire dataset
df = pd.read_csv(config.FINAL_DATASET_CSV)

all_features = []
for idx, row in df.iterrows():
    feats = extract_features(str(row['url']))
    feats['label'] = row['label']
    all_features.append(feats)

features_df = pd.DataFrame(all_features)
print(f"Features DataFrame shape: {features_df.shape}")
features_df.head()
```
