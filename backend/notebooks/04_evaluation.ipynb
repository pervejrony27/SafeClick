# Evaluation

Evaluate model performance and metrics.

```python
# Model Evaluation

import json
import sys
sys.path.insert(0, '..')
import config
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load model info
with open(config.MODEL_INFO_PATH, 'r') as f:
    info = json.load(f)

print("Model Information:")
print(f"  Type: {info['model_type']}")
print(f"  Features: {info['num_features']}")
print(f"  Training samples: {info['num_training_samples']}")
print(f"  Trained at: {info['trained_at']}")
print(f"\nMetrics:")
for key, value in info['metrics'].items():
    if key != 'confusion_matrix':
        print(f"  {key}: {value}")
```

```python
# Confusion Matrix visualization
cm = np.array(info['metrics']['confusion_matrix'])

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
            xticklabels=['Legitimate', 'Phishing'],
            yticklabels=['Legitimate', 'Phishing'])
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
ax.set_title('Confusion Matrix')
plt.tight_layout()
plt.show()
```

```python
# Test live predictions
from ml.feature_extractor import extract_features
from ml.predict import get_prediction
from ml.explainer import get_reasons

test_urls = [
    "https://www.google.com",
    "http://192.168.1.1/login/bank/verify.php",
    "http://free-prize.tk/claim-now",
    "https://www.github.com",
]

for url in test_urls:
    features = extract_features(url)
    score = get_prediction(features)
    reasons = get_reasons(url, features, score)
    
    print(f"\nURL: {url}")
    print(f"Score: {score}/100")
    print(f"Reasons: {reasons[:2]}")
    print("-" * 50)
```
